## Logistic Regression

- for housing prices and the demand of clean air

```python
import pandas as pd
```

```python
df = pd.read_csv('./dataset/housing.csv', delim_whitespace=True, header=None)
df.head()
```

```python
df.info()
```

### columns:
- 0: CRIM (crime rate)
- 1: ZN
- 2: INDUS
- 3: CHAS (1: near river, 0 : else)
- 4: NOX 
- 5: RM (averge room number)
- 6: age
- 7: DIS (distance to boston center area - employement)
- 8: RAD 
- 9: TAX
- 10: PTRATIO (teacher student ratio)
- 11: B (black population ratio)
- 12: LSTAT (low income population)
- 13: price (unit $1,000)

```python
from sklearn.model_selection import train_test_split
import numpy as np
import tensorflow as tf

seed = 0
np.random.seed(seed)
tf.set_random_seed(seed)

X = df.values[:, :13]
Y = df.values[:, 13]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=seed)
```

### model

```python
from keras.models import Sequential
from keras.layers.core import Dense

model = Sequential()
model.add(Dense(30, input_dim=13, activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1))

model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(x_train, y_train, epochs=200, batch_size=10)
```

```python
y_predict = model.predict(x_test).flatten()
for i in range(10):
    label = y_test[i]
    prediction = y_predict[i]
    print('real : {:.3f}, expected : {:.3f}'.format(label, prediction))
```
